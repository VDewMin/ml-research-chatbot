```python
#to search and download papers
import arxiv

#to handle folders and file paths
from pathlib import Path

#to add a small delay between downloads
import time

PAPERS_DIR = Path("../papers")

#create folder if it doesn't exist
PAPERS_DIR.mkdir(parents=True, exist_ok=True)

print(f"Papers will be saved to: {PAPERS_DIR.resolve()}")
```

```python
# Each entry is a tuple of (arxiv_id, short_name)
PAPERS = [
    ("1706.03762", "attention_is_all_you_need"),
    ("1810.04805", "bert"),
    ("1512.03385", "resnet"),
    ("1406.2661",  "gan"),
    ("2010.11929", "vision_transformer"),
    ("2005.11401", "rag"),
    ("1301.3666",  "word2vec"),
    ("1412.6980",  "adam_optimizer"),
    ("1502.03167", "batch_normalization"),
    ("1207.0580",  "dropout"),
    ("1509.06461", "deep_rl_atari"),
    ("2001.08361", "scaling_laws"),
]

print(f"Total papers to download: {len(PAPERS)}")
```

```python
import httpx

# Create one ArXiv client
client = arxiv.Client()

# Loop through each paper in our list
for arxiv_id, name in PAPERS:

    # Build the full file path where this PDF will be saved
    save_path = PAPERS_DIR / f"{name}.pdf"

    # If the file already exists, skip it
    if save_path.exists():
        print(f"Already exists, skipping: {name}")
        continue

    # Search ArXiv for this specific paper by its ID
    search = arxiv.Search(id_list=[arxiv_id])

    # Get the first result
    paper = next(client.results(search))

    # Build the direct PDF download URL from the ArXiv ID
    pdf_url = f"https://arxiv.org/pdf/{arxiv_id}"

    # Download the PDF content using httpx
    response = httpx.get(pdf_url, follow_redirects=True)

    # Save the raw bytes to our papers folder
    save_path.write_bytes(response.content)

    print(f"Downloaded: {name}")

    # Wait 3 seconds between downloads
    time.sleep(3)

print("\nAll papers downloaded!")
```
